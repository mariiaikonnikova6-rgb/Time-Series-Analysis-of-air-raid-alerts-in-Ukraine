```python
# ---------------------------------------------------------------------
# Dataset structure audit: geographic levels and oblast names
# ---------------------------------------------------------------------

df_raw = df_raw.copy()

print("=" * 70)
print("RECORDS BY GEOGRAPHIC LEVEL")
print("=" * 70)

display(
    df_raw["level"]
    .value_counts(dropna=False)
    .rename_axis("level")
    .rename("record_count")
    .to_frame()
)

print("\n" + "=" * 70)
print("UNIQUE OBLAST VALUES: FIRST 30 ALPHABETICALLY")
print("=" * 70)

unique_oblasts = (
    df_raw["oblast"]
    .dropna()
    .astype(str)
    .sort_values()
    .unique()
)

print(f"Total unique oblast values: {len(unique_oblasts)}")
display(pd.DataFrame({"oblast": unique_oblasts[:30]}))

print("\n" + "=" * 70)
print("KYIV-RELATED OBLAST VALUES")
print("=" * 70)

kyiv_matches = [
    oblast
    for oblast in unique_oblasts
    if "kyiv" in oblast.lower() or "kiev" in oblast.lower()
]

print(kyiv_matches if kyiv_matches else "No Kyiv-related oblast names found.")

print("\n" + "=" * 70)
print("OBLAST-LEVEL DATASET")
print("=" * 70)

df_oblast_preview = df_raw.loc[df_raw["level"] == "oblast"].copy()

print(f"Rows with level == 'oblast': {len(df_oblast_preview):,}")
print(f"Unique oblasts at oblast level: {df_oblast_preview['oblast'].nunique()}")

display(
    df_oblast_preview["oblast"]
    .value_counts()
    .sort_index()
    .rename_axis("oblast")
    .rename("record_count")
    .to_frame()
)

print("\n" + "=" * 70)
print("DATE RANGE IN RAW DATA")
print("=" * 70)

started_at_utc = pd.to_datetime(df_raw["started_at"], utc=True, errors="coerce")
finished_at_utc = pd.to_datetime(df_raw["finished_at"], utc=True, errors="coerce")

print(f"Minimum start time (UTC): {started_at_utc.min()}")
print(f"Maximum finish time (UTC): {finished_at_utc.max()}")

print("\nMissing timestamps after parsing:")
print(f"started_at:  {started_at_utc.isna().sum()}")
print(f"finished_at: {finished_at_utc.isna().sum()}")
```

```python
from config import (
    ANALYSIS_END_DATE,
    ANALYSIS_START_DATE,
    ANALYSIS_TIMEZONE,
    FORECAST_REGION,
    TARGET_GEOGRAPHIC_LEVEL,
)


# ---------------------------------------------------------------------
# Filter only oblast-level records
# ---------------------------------------------------------------------

df_oblast = (
    df_raw.loc[
        df_raw["level"].eq(TARGET_GEOGRAPHIC_LEVEL),
        ["oblast", "level", "started_at", "finished_at", "source"],
    ]
    .copy()
    .rename(columns={"oblast": "region"})
)

print("=" * 70)
print("OBLAST-LEVEL FILTER")
print("=" * 70)

print(f"Target level: {TARGET_GEOGRAPHIC_LEVEL}")
print(f"Rows after level filter: {len(df_oblast):,}")
print(f"Unique regions: {df_oblast['region'].nunique()}")

display(
    df_oblast["region"]
    .value_counts()
    .sort_index()
    .rename_axis("region")
    .rename("record_count")
    .to_frame()
)


# ---------------------------------------------------------------------
# Parse timestamps and convert UTC -> Europe/Kyiv
# ---------------------------------------------------------------------

df_oblast["started_at_utc"] = pd.to_datetime(
    df_oblast["started_at"],
    utc=True,
    errors="coerce",
)

df_oblast["finished_at_utc"] = pd.to_datetime(
    df_oblast["finished_at"],
    utc=True,
    errors="coerce",
)

df_oblast["started_at_local"] = (
    df_oblast["started_at_utc"]
    .dt.tz_convert(ANALYSIS_TIMEZONE)
)

df_oblast["finished_at_local"] = (
    df_oblast["finished_at_utc"]
    .dt.tz_convert(ANALYSIS_TIMEZONE)
)


# ---------------------------------------------------------------------
# Calculate event duration
# ---------------------------------------------------------------------

df_oblast["duration_min"] = (
    df_oblast["finished_at_utc"] - df_oblast["started_at_utc"]
).dt.total_seconds() / 60


# ---------------------------------------------------------------------
# Restrict to the selected historical analysis period
# ---------------------------------------------------------------------

analysis_start = pd.Timestamp(
    ANALYSIS_START_DATE,
    tz=ANALYSIS_TIMEZONE,
)

analysis_end_exclusive = (
    pd.Timestamp(ANALYSIS_END_DATE, tz=ANALYSIS_TIMEZONE)
    + pd.Timedelta(days=1)
)

df_analysis_preview = df_oblast.loc[
    df_oblast["started_at_local"].between(
        analysis_start,
        analysis_end_exclusive,
        inclusive="left",
    )
].copy()


# ---------------------------------------------------------------------
# Quality checks: not removing anything yet
# ---------------------------------------------------------------------

duplicate_event_count = df_analysis_preview.duplicated(
    subset=["region", "started_at_utc", "finished_at_utc"],
    keep=False,
).sum()

invalid_duration_count = (
    df_analysis_preview["duration_min"].le(0).sum()
)

missing_timestamp_count = (
    df_analysis_preview["started_at_utc"].isna().sum()
    + df_analysis_preview["finished_at_utc"].isna().sum()
)

print("\n" + "=" * 70)
print("ANALYSIS PERIOD PREVIEW")
print("=" * 70)

print(f"Analysis start: {analysis_start}")
print(f"Analysis end:   {analysis_end_exclusive}")
print(f"Rows in selected period: {len(df_analysis_preview):,}")
print(f"Regions in selected period: {df_analysis_preview['region'].nunique()}")

print("\n" + "=" * 70)
print("DATA QUALITY PREVIEW")
print("=" * 70)

print(f"Missing parsed timestamps: {missing_timestamp_count:,}")
print(f"Events with duration <= 0 minutes: {invalid_duration_count:,}")
print(f"Rows belonging to exact duplicate events: {duplicate_event_count:,}")

print("\nDuration summary, minutes:")
display(
    df_analysis_preview["duration_min"]
    .describe()
    .to_frame(name="duration_min")
)

print("\n" + "=" * 70)
print("FORECAST REGION CHECK")
print("=" * 70)

df_kyiv_preview = df_analysis_preview.loc[
    df_analysis_preview["region"].eq(FORECAST_REGION)
].copy()

print(f"Forecast region: {FORECAST_REGION}")
print(f"Number of events: {len(df_kyiv_preview):,}")
print(
    f"Date range: "
    f"{df_kyiv_preview['started_at_local'].min()} "
    f"to {df_kyiv_preview['finished_at_local'].max()}"
)

display(
    df_kyiv_preview[
        [
            "region",
            "started_at_local",
            "finished_at_local",
            "duration_min",
        ]
    ].head(10)
)
```

```python
# =====================================================================
# 01_data_audit.ipynb
# Historical Time Series Analysis of Air Raid Alerts in Ukraine
# Step 1: Initial data loading and audit
# =====================================================================

from pathlib import Path
import sys
import importlib

import pandas as pd
from IPython.display import display


# ---------------------------------------------------------------------
# 1. Find the project root and the src/ directory
# ---------------------------------------------------------------------

current_dir = Path.cwd().resolve()

# The notebook is expected to run either from:
# - project_root/
# - project_root/notebooks/
if (current_dir / "src").exists():
    PROJECT_ROOT = current_dir

elif (current_dir.parent / "src").exists():
    PROJECT_ROOT = current_dir.parent

else:
    raise FileNotFoundError(
        "Project root was not found.\n"
        "Expected to find a folder named 'src' in the current directory "
        "or in its parent directory.\n"
        f"Current working directory: {current_dir}"
    )

SRC_DIR = PROJECT_ROOT / "src"
CONFIG_FILE = SRC_DIR / "config.py"

print("=" * 78)
print("PROJECT PATH CHECK")
print("=" * 78)

print(f"Current working directory: {current_dir}")
print(f"Detected project root:    {PROJECT_ROOT}")
print(f"Detected src directory:   {SRC_DIR}")
print(f"Expected config.py path:  {CONFIG_FILE}")
print(f"Does src directory exist? {SRC_DIR.exists()}")
print(f"Does config.py exist?     {CONFIG_FILE.exists()}")

if not CONFIG_FILE.exists():
    raise FileNotFoundError(
        "config.py was not found.\n"
        f"Expected path: {CONFIG_FILE}"
    )


# ---------------------------------------------------------------------
# 2. Add src/ to Python import paths and reload config.py
# ---------------------------------------------------------------------

if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

# Remove a possible old version of config from Jupyter memory.
# This is useful when config.py was edited during notebook development.
if "config" in sys.modules:
    del sys.modules["config"]

importlib.invalidate_caches()

import config

print("\n" + "=" * 78)
print("CONFIG CHECK")
print("=" * 78)

print(f"Loaded config module from: {config.__file__}")

required_constants = [
    "RAW_ALERTS_FILE",
    "ANALYSIS_START_DATE",
    "ANALYSIS_END_DATE",
    "ANALYSIS_TIMEZONE",
    "TARGET_GEOGRAPHIC_LEVEL",
    "FORECAST_REGION",
    "SELECTED_REGIONS",
]

missing_constants = []

for constant_name in required_constants:
    if hasattr(config, constant_name):
        print(f"{constant_name}: {getattr(config, constant_name)}")
    else:
        missing_constants.append(constant_name)
        print(f"{constant_name}: NOT FOUND")

if missing_constants:
    raise AttributeError(
        "The following constants are missing in src/config.py:\n"
        f"{missing_constants}"
    )

# Copy values from config.py into notebook variables.
RAW_ALERTS_FILE = config.RAW_ALERTS_FILE
ANALYSIS_START_DATE = config.ANALYSIS_START_DATE
ANALYSIS_END_DATE = config.ANALYSIS_END_DATE
ANALYSIS_TIMEZONE = config.ANALYSIS_TIMEZONE
TARGET_GEOGRAPHIC_LEVEL = config.TARGET_GEOGRAPHIC_LEVEL
FORECAST_REGION = config.FORECAST_REGION
SELECTED_REGIONS = config.SELECTED_REGIONS


# ---------------------------------------------------------------------
# 3. Load the raw CSV dataset
# ---------------------------------------------------------------------

print("\n" + "=" * 78)
print("RAW DATASET LOADING")
print("=" * 78)

print(f"Expected dataset path: {RAW_ALERTS_FILE}")

try:
    df_raw = pd.read_csv(RAW_ALERTS_FILE)

except FileNotFoundError:
    print("\n❌ Dataset file was not found.")
    print("Save the file at this path:")
    print(f"   {RAW_ALERTS_FILE}")
    raise

except pd.errors.EmptyDataError:
    print("\n❌ The dataset file exists, but it is empty.")
    raise

except pd.errors.ParserError as error:
    print("\n❌ CSV parsing error:")
    print(error)
    raise

else:
    print("\n✅ Dataset loaded successfully.")
    print(f"Shape: {df_raw.shape[0]:,} rows × {df_raw.shape[1]} columns")


# ---------------------------------------------------------------------
# 4. Basic audit of raw data
# ---------------------------------------------------------------------

print("\n" + "=" * 78)
print("RAW DATAFRAME INFO")
print("=" * 78)

df_raw.info()

print("\n" + "=" * 78)
print("FIRST 5 ROWS")
print("=" * 78)

display(df_raw.head())

print("\n" + "=" * 78)
print("MISSING VALUES BY COLUMN")
print("=" * 78)

display(
    df_raw.isnull()
    .sum()
    .sort_values(ascending=False)
    .rename("missing_values")
    .to_frame()
)

print("\n" + "=" * 78)
print("RAW DATA COLUMNS")
print("=" * 78)

print(df_raw.columns.tolist())


# ---------------------------------------------------------------------
# 5. Validate that expected source columns are present
# ---------------------------------------------------------------------

required_raw_columns = [
    "oblast",
    "raion",
    "hromada",
    "level",
    "started_at",
    "finished_at",
    "source",
]

missing_raw_columns = [
    column
    for column in required_raw_columns
    if column not in df_raw.columns
]

if missing_raw_columns:
    raise KeyError(
        "The dataset does not contain expected columns:\n"
        f"{missing_raw_columns}\n\n"
        f"Available columns:\n{df_raw.columns.tolist()}"
    )


# ---------------------------------------------------------------------
# 6. Audit geographic levels
# ---------------------------------------------------------------------

print("\n" + "=" * 78)
print("RECORDS BY GEOGRAPHIC LEVEL")
print("=" * 78)

display(
    df_raw["level"]
    .value_counts(dropna=False)
    .rename_axis("level")
    .rename("record_count")
    .to_frame()
)

print("\n" + "=" * 78)
print("UNIQUE OBLAST VALUES: ALPHABETICALLY")
print("=" * 78)

unique_oblasts = (
    df_raw["oblast"]
    .dropna()
    .astype(str)
    .sort_values()
    .unique()
)

print(f"Total unique oblast values: {len(unique_oblasts)}")

display(
    pd.DataFrame(
        {
            "oblast": unique_oblasts
        }
    )
)

print("\n" + "=" * 78)
print("KYIV-RELATED OBLAST VALUES")
print("=" * 78)

kyiv_matches = [
    oblast
    for oblast in unique_oblasts
    if "kyiv" in oblast.lower() or "kiev" in oblast.lower()
]

print(kyiv_matches if kyiv_matches else "No Kyiv-related oblast names found.")


# ---------------------------------------------------------------------
# 7. Filter only oblast-level records
# ---------------------------------------------------------------------

df_oblast = (
    df_raw.loc[
        df_raw["level"].eq(TARGET_GEOGRAPHIC_LEVEL),
        [
            "oblast",
            "level",
            "started_at",
            "finished_at",
            "source",
        ],
    ]
    .copy()
    .rename(columns={"oblast": "region"})
)

print("\n" + "=" * 78)
print("OBLAST-LEVEL DATASET")
print("=" * 78)

print(f"Target geographic level: {TARGET_GEOGRAPHIC_LEVEL}")
print(f"Rows after level filter: {len(df_oblast):,}")
print(f"Unique regions: {df_oblast['region'].nunique()}")

display(
    df_oblast["region"]
    .value_counts()
    .sort_index()
    .rename_axis("region")
    .rename("record_count")
    .to_frame()
)


# ---------------------------------------------------------------------
# 8. Parse timestamps and convert UTC -> Europe/Kyiv
# ---------------------------------------------------------------------

df_oblast["started_at_utc"] = pd.to_datetime(
    df_oblast["started_at"],
    utc=True,
    errors="coerce",
)

df_oblast["finished_at_utc"] = pd.to_datetime(
    df_oblast["finished_at"],
    utc=True,
    errors="coerce",
)

df_oblast["started_at_local"] = (
    df_oblast["started_at_utc"]
    .dt.tz_convert(ANALYSIS_TIMEZONE)
)

df_oblast["finished_at_local"] = (
    df_oblast["finished_at_utc"]
    .dt.tz_convert(ANALYSIS_TIMEZONE)
)


# ---------------------------------------------------------------------
# 9. Calculate duration for every alert event
# ---------------------------------------------------------------------

df_oblast["duration_min"] = (
    df_oblast["finished_at_utc"]
    - df_oblast["started_at_utc"]
).dt.total_seconds() / 60


# ---------------------------------------------------------------------
# 10. Restrict the data to the MVP analysis period
# ---------------------------------------------------------------------

analysis_start = pd.Timestamp(
    ANALYSIS_START_DATE,
    tz=ANALYSIS_TIMEZONE,
)

analysis_end_exclusive = (
    pd.Timestamp(
        ANALYSIS_END_DATE,
        tz=ANALYSIS_TIMEZONE,
    )
    + pd.Timedelta(days=1)
)

period_mask = (
    df_oblast["started_at_local"].ge(analysis_start)
    & df_oblast["started_at_local"].lt(analysis_end_exclusive)
)

df_analysis_preview = df_oblast.loc[period_mask].copy()

print("\n" + "=" * 78)
print("ANALYSIS PERIOD PREVIEW")
print("=" * 78)

print(f"Analysis start: {analysis_start}")
print(f"Analysis end:   {analysis_end_exclusive}")
print(f"Rows in selected period: {len(df_analysis_preview):,}")
print(f"Regions in selected period: {df_analysis_preview['region'].nunique()}")

print("\nDate range after filtering:")
print(f"Minimum local start: {df_analysis_preview['started_at_local'].min()}")
print(f"Maximum local finish: {df_analysis_preview['finished_at_local'].max()}")


# ---------------------------------------------------------------------
# 11. Primary data-quality checks
# ---------------------------------------------------------------------

missing_started_at_count = df_analysis_preview["started_at_utc"].isna().sum()
missing_finished_at_count = df_analysis_preview["finished_at_utc"].isna().sum()

invalid_duration_count = (
    df_analysis_preview["duration_min"]
    .le(0)
    .sum()
)

duplicate_event_rows_count = (
    df_analysis_preview.duplicated(
        subset=[
            "region",
            "started_at_utc",
            "finished_at_utc",
        ],
        keep=False,
    )
    .sum()
)

print("\n" + "=" * 78)
print("DATA QUALITY PREVIEW")
print("=" * 78)

print(f"Missing parsed started_at timestamps:  {missing_started_at_count:,}")
print(f"Missing parsed finished_at timestamps: {missing_finished_at_count:,}")
print(f"Events with duration <= 0 minutes:     {invalid_duration_count:,}")
print(f"Rows belonging to exact duplicates:     {duplicate_event_rows_count:,}")

print("\nDuration summary, minutes:")
display(
    df_analysis_preview["duration_min"]
    .describe()
    .to_frame(name="duration_min")
)


# ---------------------------------------------------------------------
# 12. Check the selected regions from config.py
# ---------------------------------------------------------------------

print("\n" + "=" * 78)
print("SELECTED REGIONS CHECK")
print("=" * 78)

available_regions = set(df_analysis_preview["region"].unique())

selected_regions_status = pd.DataFrame(
    {
        "region": SELECTED_REGIONS,
        "exists_in_analysis_data": [
            region in available_regions
            for region in SELECTED_REGIONS
        ],
    }
)

display(selected_regions_status)


# ---------------------------------------------------------------------
# 13. Check the forecast region: Kyiv City
# ---------------------------------------------------------------------

df_kyiv_preview = df_analysis_preview.loc[
    df_analysis_preview["region"].eq(FORECAST_REGION)
].copy()

print("\n" + "=" * 78)
print("FORECAST REGION CHECK")
print("=" * 78)

print(f"Forecast region: {FORECAST_REGION}")
print(f"Number of events: {len(df_kyiv_preview):,}")

if df_kyiv_preview.empty:
    print("\n⚠️ No records were found for the forecast region.")
    print("Check FORECAST_REGION in src/config.py.")

else:
    print(
        "Date range:\n"
        f"  Start: {df_kyiv_preview['started_at_local'].min()}\n"
        f"  End:   {df_kyiv_preview['finished_at_local'].max()}"
    )

    display(
        df_kyiv_preview[
            [
                "region",
                "started_at_local",
                "finished_at_local",
                "duration_min",
            ]
        ].head(10)
    )


# ---------------------------------------------------------------------
# 14. Keep a compact working preview for the next notebook steps
# ---------------------------------------------------------------------

print("\n" + "=" * 78)
print("WORKING DATAFRAME PREVIEW")
print("=" * 78)

display(
    df_analysis_preview[
        [
            "region",
            "started_at_utc",
            "finished_at_utc",
            "started_at_local",
            "finished_at_local",
            "duration_min",
        ]
    ].head()
)


# =====================================================================
# Duplicate-event investigation
# =====================================================================

duplicate_key_columns = [
    "region",
    "started_at_utc",
    "finished_at_utc",
]

# Розмір кожної групи подій із однаковими region/start/end.
event_group_sizes = (
    df_analysis_preview
    .groupby(duplicate_key_columns, dropna=False)
    .size()
    .rename("copies")
)

print("=" * 78)
print("DUPLICATE EVENT INVESTIGATION")
print("=" * 78)

print(f"Total rows in analysis dataset: {len(df_analysis_preview):,}")
print(f"Unique event groups:            {len(event_group_sizes):,}")
print(
    f"Extra duplicate rows:            "
    f"{len(df_analysis_preview) - len(event_group_sizes):,}"
)

print("\nDistribution of copies per event:")
display(
    event_group_sizes
    .value_counts()
    .sort_index()
    .rename_axis("copies_per_event")
    .rename("number_of_event_groups")
    .to_frame()
)

# Отримуємо лише події, які повторюються.
duplicate_groups = (
    event_group_sizes[event_group_sizes > 1]
    .reset_index()
    .sort_values(
        by=["copies", "region", "started_at_utc"],
        ascending=[False, True, True],
    )
)

print("\nNumber of duplicated event groups:")
print(f"{len(duplicate_groups):,}")

print("\nFirst 10 duplicated event keys:")
display(duplicate_groups.head(10))

# Вибираємо перші кілька дублікатів для ручної перевірки.
sample_duplicate_keys = duplicate_groups.head(5)[duplicate_key_columns]

sample_duplicate_rows = (
    df_analysis_preview
    .merge(
        sample_duplicate_keys,
        on=duplicate_key_columns,
        how="inner",
    )
    .sort_values(duplicate_key_columns)
)

print("\n" + "=" * 78)
print("SAMPLE DUPLICATE ROWS FROM WORKING DATASET")
print("=" * 78)

display(
    sample_duplicate_rows[
        [
            "region",
            "level",
            "started_at",
            "finished_at",
            "source",
            "started_at_utc",
            "finished_at_utc",
            "duration_min",
        ]
    ]
)

# Перевіряємо оригінальні колонки raw CSV для тих самих рядків.
# Це допоможе встановити, чи дублікати відрізнялися raion або hromada.
sample_original_indices = sample_duplicate_rows.index.unique()

print("\n" + "=" * 78)
print("SAME ROWS IN ORIGINAL RAW DATASET")
print("=" * 78)

display(
    df_raw.loc[
        sample_original_indices,
        [
            "oblast",
            "raion",
            "hromada",
            "level",
            "started_at",
            "finished_at",
            "source",
        ],
    ]
)


df_clean = df_analysis_preview.drop_duplicates(
    subset=[
        "region",
        "started_at_utc",
        "finished_at_utc",
    ]
).copy()
```

```python
# =====================================================================
# Duplicate-event investigation
# =====================================================================

duplicate_key_columns = [
    "region",
    "started_at_utc",
    "finished_at_utc",
]

# Розмір кожної групи подій із однаковими region/start/end.
event_group_sizes = (
    df_analysis_preview
    .groupby(duplicate_key_columns, dropna=False)
    .size()
    .rename("copies")
)

print("=" * 78)
print("DUPLICATE EVENT INVESTIGATION")
print("=" * 78)

print(f"Total rows in analysis dataset: {len(df_analysis_preview):,}")
print(f"Unique event groups:            {len(event_group_sizes):,}")
print(
    f"Extra duplicate rows:            "
    f"{len(df_analysis_preview) - len(event_group_sizes):,}"
)

print("\nDistribution of copies per event:")
display(
    event_group_sizes
    .value_counts()
    .sort_index()
    .rename_axis("copies_per_event")
    .rename("number_of_event_groups")
    .to_frame()
)

# Отримуємо лише події, які повторюються.
duplicate_groups = (
    event_group_sizes[event_group_sizes > 1]
    .reset_index()
    .sort_values(
        by=["copies", "region", "started_at_utc"],
        ascending=[False, True, True],
    )
)

print("\nNumber of duplicated event groups:")
print(f"{len(duplicate_groups):,}")

print("\nFirst 10 duplicated event keys:")
display(duplicate_groups.head(10))

# Вибираємо перші кілька дублікатів для ручної перевірки.
sample_duplicate_keys = duplicate_groups.head(5)[duplicate_key_columns]

sample_duplicate_rows = (
    df_analysis_preview
    .merge(
        sample_duplicate_keys,
        on=duplicate_key_columns,
        how="inner",
    )
    .sort_values(duplicate_key_columns)
)

print("\n" + "=" * 78)
print("SAMPLE DUPLICATE ROWS FROM WORKING DATASET")
print("=" * 78)

display(
    sample_duplicate_rows[
        [
            "region",
            "level",
            "started_at",
            "finished_at",
            "source",
            "started_at_utc",
            "finished_at_utc",
            "duration_min",
        ]
    ]
)

# Перевіряємо оригінальні колонки raw CSV для тих самих рядків.
# Це допоможе встановити, чи дублікати відрізнялися raion або hromada.
sample_original_indices = sample_duplicate_rows.index.unique()

print("\n" + "=" * 78)
print("SAME ROWS IN ORIGINAL RAW DATASET")
print("=" * 78)

display(
    df_raw.loc[
        sample_original_indices,
        [
            "oblast",
            "raion",
            "hromada",
            "level",
            "started_at",
            "finished_at",
            "source",
        ],
    ]
)

df_clean = df_analysis_preview.drop_duplicates(
    subset=[
        "region",
        "started_at_utc",
        "finished_at_utc",
    ]
).copy()
```

```python
# =====================================================================
# Duplicate-event investigation
# =====================================================================

duplicate_key_columns = [
    "region",
    "started_at_utc",
    "finished_at_utc",
]

event_group_sizes = (
    df_analysis_preview
    .groupby(duplicate_key_columns, dropna=False)
    .size()
    .rename("copies")
)

print("=" * 78)
print("DUPLICATE EVENT INVESTIGATION")
print("=" * 78)

print(f"Total rows in analysis dataset: {len(df_analysis_preview):,}")
print(f"Unique event groups:            {len(event_group_sizes):,}")
print(
    f"Extra duplicate rows:            "
    f"{len(df_analysis_preview) - len(event_group_sizes):,}"
)

print("\nDistribution of copies per event:")
display(
    event_group_sizes
    .value_counts()
    .sort_index()
    .rename_axis("copies_per_event")
    .rename("number_of_event_groups")
    .to_frame()
)

duplicate_groups = (
    event_group_sizes[event_group_sizes > 1]
    .reset_index()
    .sort_values(
        by=["copies", "region", "started_at_utc"],
        ascending=[False, True, True],
    )
)

print("\nNumber of duplicated event groups:")
print(f"{len(duplicate_groups):,}")

print("\nFirst 10 duplicated event keys:")
display(duplicate_groups.head(10))

sample_duplicate_keys = duplicate_groups.head(5)[duplicate_key_columns]

sample_duplicate_rows = (
    df_analysis_preview
    .reset_index(names="original_index")
    .merge(
        sample_duplicate_keys,
        on=duplicate_key_columns,
        how="inner",
    )
    .sort_values(duplicate_key_columns)
)

print("\n" + "=" * 78)
print("SAMPLE DUPLICATE ROWS FROM WORKING DATASET")
print("=" * 78)

display(
    sample_duplicate_rows[
        [
            "original_index",
            "region",
            "level",
            "started_at",
            "finished_at",
            "source",
            "started_at_utc",
            "finished_at_utc",
            "duration_min",
        ]
    ]
)

print("\n" + "=" * 78)
print("SAME ROWS IN ORIGINAL RAW DATASET")
print("=" * 78)

sample_original_indices = sample_duplicate_rows["original_index"].unique()

display(
    df_raw.loc[
        sample_original_indices,
        [
            "oblast",
            "raion",
            "hromada",
            "level",
            "started_at",
            "finished_at",
            "source",
        ],
    ]
)

# =====================================================================
# Deduplication preview: one row = one oblast-level alert event
# =====================================================================

EVENT_KEY_COLUMNS = [
    "region",
    "started_at_utc",
    "finished_at_utc",
    "source",
]

rows_before_deduplication = len(df_analysis_preview)

# Count groups before removing duplicate event copies.
event_group_sizes_check = (
    df_analysis_preview
    .groupby(EVENT_KEY_COLUMNS, dropna=False)
    .size()
    .rename("copies")
)

# Keep the first occurrence of each oblast-level event.
df_cleaned_preview = (
    df_analysis_preview
    .drop_duplicates(
        subset=EVENT_KEY_COLUMNS,
        keep="first",
    )
    .sort_values(
        by=["region", "started_at_utc", "finished_at_utc"],
    )
    .reset_index(drop=True)
)

rows_after_deduplication = len(df_cleaned_preview)
rows_removed_as_duplicates = (
    rows_before_deduplication - rows_after_deduplication
)

remaining_duplicate_rows = (
    df_cleaned_preview
    .duplicated(
        subset=EVENT_KEY_COLUMNS,
        keep=False,
    )
    .sum()
)

kyiv_events_before = (
    df_analysis_preview["region"]
    .eq(FORECAST_REGION)
    .sum()
)

kyiv_events_after = (
    df_cleaned_preview["region"]
    .eq(FORECAST_REGION)
    .sum()
)

print("=" * 78)
print("DEDUPLICATION PREVIEW")
print("=" * 78)

print(f"Rows before removing duplicates: {rows_before_deduplication:,}")
print(f"Rows after removing duplicates:  {rows_after_deduplication:,}")
print(f"Rows removed as duplicates:      {rows_removed_as_duplicates:,}")
print(f"Remaining duplicated event rows: {remaining_duplicate_rows:,}")

print("\nDuplicate-group distribution before cleaning:")
display(
    event_group_sizes_check
    .value_counts()
    .sort_index()
    .rename_axis("copies_per_event")
    .rename("number_of_event_groups")
    .to_frame()
)

print("\n" + "=" * 78)
print("FORECAST REGION AFTER DEDUPLICATION")
print("=" * 78)

print(f"Forecast region: {FORECAST_REGION}")
print(f"Events before deduplication: {kyiv_events_before:,}")
print(f"Events after deduplication:  {kyiv_events_after:,}")

print("\n" + "=" * 78)
print("CLEANED DATA PREVIEW")
print("=" * 78)

display(
    df_cleaned_preview[
        [
            "region",
            "started_at_local",
            "finished_at_local",
            "duration_min",
            "source",
        ]
    ].head(10)
)

print("\n" + "=" * 78)
print("CLEANING ASSERTIONS")
print("=" * 78)

assert rows_after_deduplication == len(event_group_sizes_check), (
    "The number of rows after deduplication does not match "
    "the number of unique event groups."
)

assert remaining_duplicate_rows == 0, (
    "Duplicate event rows still remain after deduplication."
)

assert (df_cleaned_preview["duration_min"] > 0).all(), (
    "Non-positive durations are present in the cleaned dataset."
)

print("✅ All cleaning checks passed successfully.")
```

```python
# =====================================================================
# Daily aggregation preview
# Split alert events across local calendar days and create daily metrics
# =====================================================================

from datetime import timedelta

import numpy as np
import pandas as pd
from IPython.display import display


# ---------------------------------------------------------------------
# 1. Check that the cleaned dataset from preprocessing exists
# ---------------------------------------------------------------------

required_variables = [
    "df_cleaned",
    "ANALYSIS_START_DATE",
    "ANALYSIS_END_DATE",
    "ANALYSIS_TIMEZONE",
    "FORECAST_REGION",
]

missing_variables = [
    variable_name
    for variable_name in required_variables
    if variable_name not in globals()
]

if missing_variables:
    raise NameError(
        "The following variables are missing from the notebook session:\n"
        f"{missing_variables}\n\n"
        "Run the preprocessing cell that creates df_cleaned first."
    )


# ---------------------------------------------------------------------
# 2. Define local analysis-period boundaries
# ---------------------------------------------------------------------

analysis_start_local = pd.Timestamp(
    ANALYSIS_START_DATE,
    tz=ANALYSIS_TIMEZONE,
)

# The end boundary is exclusive:
# 2025-12-01 00:00:00 local time is not included.
analysis_end_exclusive_local = (
    pd.Timestamp(
        ANALYSIS_END_DATE,
        tz=ANALYSIS_TIMEZONE,
    )
    + pd.DateOffset(days=1)
)

print("=" * 78)
print("DAILY AGGREGATION SETTINGS")
print("=" * 78)

print(f"Analysis start (local): {analysis_start_local}")
print(f"Analysis end (exclusive, local): {analysis_end_exclusive_local}")
print(f"Input cleaned events: {len(df_cleaned):,}")


# ---------------------------------------------------------------------
# 3. Prepare event-level data and create stable event IDs
# ---------------------------------------------------------------------

events_for_aggregation = (
    df_cleaned[
        [
            "region",
            "started_at_local",
            "finished_at_local",
            "duration_min",
        ]
    ]
    .copy()
    .reset_index(drop=True)
)

events_for_aggregation.insert(
    loc=0,
    column="event_id",
    value=events_for_aggregation.index,
)

# Count events that cross a local midnight.
cross_midnight_events = (
    events_for_aggregation["started_at_local"].dt.date
    != events_for_aggregation["finished_at_local"].dt.date
).sum()

print(f"Events spanning local midnight: {cross_midnight_events:,}")


# ---------------------------------------------------------------------
# 4. Split each event into separate local calendar-day segments
# ---------------------------------------------------------------------

def split_event_into_daily_segments(
    event_id: int,
    region: str,
    started_at_local: pd.Timestamp,
    finished_at_local: pd.Timestamp,
    period_start: pd.Timestamp,
    period_end_exclusive: pd.Timestamp,
    timezone_name: str,
) -> list[dict]:
    """
    Split one alert event into one or more pieces, one per local date.

    The event is clipped to the selected analysis period first.
    This avoids accidentally assigning time after 2025-11-30
    to the MVP analysis.
    """
    clipped_start = max(started_at_local, period_start)
    clipped_end = min(finished_at_local, period_end_exclusive)

    # Ignore events that do not overlap the selected period.
    if clipped_end <= clipped_start:
        return []

    segment_rows = []

    current_date = clipped_start.date()

    while True:
        day_start = pd.Timestamp(
            current_date,
            tz=timezone_name,
        )

        next_day_start = pd.Timestamp(
            current_date + timedelta(days=1),
            tz=timezone_name,
        )

        segment_start = max(clipped_start, day_start)
        segment_end = min(clipped_end, next_day_start)

        if segment_end > segment_start:
            segment_duration_min = (
                segment_end - segment_start
            ).total_seconds() / 60

            segment_rows.append(
                {
                    "event_id": event_id,
                    "region": region,
                    "date": pd.Timestamp(current_date),
                    "segment_start_local": segment_start,
                    "segment_end_local": segment_end,
                    "segment_duration_min": segment_duration_min,
                }
            )

        # The full event was allocated across daily segments.
        if next_day_start >= clipped_end:
            break

        current_date = current_date + timedelta(days=1)

    return segment_rows


all_segments = []

for event in events_for_aggregation.itertuples(index=False):
    event_segments = split_event_into_daily_segments(
        event_id=event.event_id,
        region=event.region,
        started_at_local=event.started_at_local,
        finished_at_local=event.finished_at_local,
        period_start=analysis_start_local,
        period_end_exclusive=analysis_end_exclusive_local,
        timezone_name=ANALYSIS_TIMEZONE,
    )

    all_segments.extend(event_segments)

event_day_segments = pd.DataFrame(all_segments)

if event_day_segments.empty:
    raise ValueError(
        "No daily segments were created. "
        "Check the date range and timestamp columns."
    )


# ---------------------------------------------------------------------
# 5. Aggregate duration and active events by region and local date
# ---------------------------------------------------------------------

daily_duration_metrics = (
    event_day_segments
    .groupby(
        ["date", "region"],
        as_index=False,
    )
    .agg(
        total_duration_min=(
            "segment_duration_min",
            "sum",
        ),
        active_alert_count=(
            "event_id",
            "nunique",
        ),
    )
)

# Count alert starts on each local day.
# This differs from active_alert_count:
# an event can be active after midnight even if it started on the day before.
daily_start_metrics = (
    events_for_aggregation
    .loc[
        events_for_aggregation["started_at_local"].ge(
            analysis_start_local
        )
        & events_for_aggregation["started_at_local"].lt(
            analysis_end_exclusive_local
        )
    ]
    .assign(
        date=lambda dataframe: (
            dataframe["started_at_local"]
            .dt.tz_localize(None)
            .dt.normalize()
        )
    )
    .groupby(
        ["date", "region"],
        as_index=False,
    )
    .agg(
        alert_start_count=(
            "event_id",
            "nunique",
        )
    )
)


# ---------------------------------------------------------------------
# 6. Create a complete date × region grid
# Days without alerts must explicitly contain zeros.
# ---------------------------------------------------------------------

analysis_dates = pd.date_range(
    start=pd.Timestamp(ANALYSIS_START_DATE),
    end=pd.Timestamp(ANALYSIS_END_DATE),
    freq="D",
)

all_regions = sorted(
    events_for_aggregation["region"].unique()
)

full_daily_grid = (
    pd.MultiIndex.from_product(
        [analysis_dates, all_regions],
        names=["date", "region"],
    )
    .to_frame(index=False)
)

daily_metrics_preview = (
    full_daily_grid
    .merge(
        daily_duration_metrics,
        on=["date", "region"],
        how="left",
    )
    .merge(
        daily_start_metrics,
        on=["date", "region"],
        how="left",
    )
)

daily_metrics_preview["total_duration_min"] = (
    daily_metrics_preview["total_duration_min"]
    .fillna(0.0)
    .astype(float)
)

daily_metrics_preview["active_alert_count"] = (
    daily_metrics_preview["active_alert_count"]
    .fillna(0)
    .astype(int)
)

daily_metrics_preview["alert_start_count"] = (
    daily_metrics_preview["alert_start_count"]
    .fillna(0)
    .astype(int)
)

daily_metrics_preview["mean_active_duration_min"] = np.where(
    daily_metrics_preview["active_alert_count"] > 0,
    daily_metrics_preview["total_duration_min"]
    / daily_metrics_preview["active_alert_count"],
    0.0,
)

daily_metrics_preview = (
    daily_metrics_preview
    .sort_values(["region", "date"])
    .reset_index(drop=True)
)


# ---------------------------------------------------------------------
# 7. Validate that duration was conserved during splitting
# ---------------------------------------------------------------------

clipped_start = events_for_aggregation["started_at_local"].where(
    events_for_aggregation["started_at_local"].ge(
        analysis_start_local
    ),
    analysis_start_local,
)

clipped_end = events_for_aggregation["finished_at_local"].where(
    events_for_aggregation["finished_at_local"].lt(
        analysis_end_exclusive_local
    ),
    analysis_end_exclusive_local,
)

expected_duration_min = (
    (clipped_end - clipped_start)
    .dt.total_seconds()
    .div(60)
    .clip(lower=0)
    .sum()
)

allocated_duration_min = (
    event_day_segments["segment_duration_min"]
    .sum()
)

duration_difference_min = (
    allocated_duration_min
    - expected_duration_min
)


# ---------------------------------------------------------------------
# 8. Quality checks
# ---------------------------------------------------------------------

duplicate_daily_rows = (
    daily_metrics_preview
    .duplicated(
        subset=["date", "region"],
        keep=False,
    )
    .sum()
)

number_of_expected_daily_rows = (
    len(analysis_dates)
    * len(all_regions)
)

print("\n" + "=" * 78)
print("DAILY AGGREGATION QUALITY CHECKS")
print("=" * 78)

print(f"Event-day segments created: {len(event_day_segments):,}")
print(f"Daily metrics rows: {len(daily_metrics_preview):,}")
print(f"Expected complete daily rows: {number_of_expected_daily_rows:,}")
print(f"Duplicate date-region rows: {duplicate_daily_rows:,}")

print(
    "\nTotal duration expected after clipping: "
    f"{expected_duration_min:,.6f} minutes"
)

print(
    "Total duration allocated to daily segments: "
    f"{allocated_duration_min:,.6f} minutes"
)

print(
    "Difference after splitting: "
    f"{duration_difference_min:.12f} minutes"
)

assert len(daily_metrics_preview) == number_of_expected_daily_rows, (
    "The daily grid is incomplete."
)

assert duplicate_daily_rows == 0, (
    "Duplicate date-region rows exist after aggregation."
)

assert np.isclose(
    allocated_duration_min,
    expected_duration_min,
    rtol=0,
    atol=1e-8,
), (
    "Duration was not conserved while splitting events "
    "across local calendar days."
)

assert (daily_metrics_preview["total_duration_min"] >= 0).all(), (
    "Negative daily durations were found."
)

print("\n✅ Daily aggregation checks passed successfully.")


# ---------------------------------------------------------------------
# 9. Display results
# ---------------------------------------------------------------------

print("\n" + "=" * 78)
print("DAILY METRICS PREVIEW")
print("=" * 78)

display(daily_metrics_preview.head(15))

print("\n" + "=" * 78)
print("KYIV CITY DAILY METRICS: FIRST 15 DAYS")
print("=" * 78)

display(
    daily_metrics_preview.loc[
        daily_metrics_preview["region"].eq(FORECAST_REGION),
        [
            "date",
            "region",
            "alert_start_count",
            "active_alert_count",
            "total_duration_min",
            "mean_active_duration_min",
        ],
    ]
    .head(15)
)

print("\n" + "=" * 78)
print("SAMPLE EVENTS THAT SPAN LOCAL MIDNIGHT")
print("=" * 78)

display(
    events_for_aggregation.loc[
        events_for_aggregation["started_at_local"].dt.date
        != events_for_aggregation["finished_at_local"].dt.date,
        [
            "event_id",
            "region",
            "started_at_local",
            "finished_at_local",
            "duration_min",
        ],
    ]
    .head(10)
)
```
